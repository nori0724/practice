### 3.3.3 描画オプション

```r
plot(georgia, col="red", bg="wheat")
```

```r
# 群の境界を結合する
georgia.outline <- gUnaryUnion(georgia, id=NULL)

# 地図をplotする
plot(georgia, col="red", bg="wheat", lty=2, border ="blue") #データセットをplotしている
plot(georgia.outline, lwd=3, add=TRUE) # データセットを1つにまとめたデータセットを1つ目の描画結果にオーバーレイしている

# タイトルを追加
title(main="The State of Georgia", font.main=2, cex.main=1.5, sub="and its counties", font.sub=3, col.sub="blue")
```

```r
# 座標を設定
Lat <- data.frame(georgia)[, 1]
Lon <- data.frame(georgia)[, 2]

# ラベルを設定
Names <- data.frame(georgia)[, 13]

# plotパラメータを設定
par(mar=c(0,0,0,0))
plot(georgia, col=NA)
pl <- pointLabel(Lon, Lat, Names, offset=0, cex=0.5)
```

```r
# ジョージア郡の東部に興味があるときのplot
country.tmp <- c(81, 82, 83, 150, 62, 53, 21, 16, 124, 121, 17)
georgia.sub <- georgia[country.tmp, ]
par(mar=c(0,0,3,0))
plot(georgia.sub, col="gold1")
plot(georgia.outline, add=TRUE, lwd=2)
title("A subset of Geirgia", cex.main=2, font.main=1)
pl <- pointLabel(Lon[country.tmp], Lat[country.tmp], Names[country.tmp], offset=3, cex=1.5)
```

```r
plot(georgia, border="grey", lwd=0.5)
plot(georgia.sub, add=TRUE, col="lightblue")
plot(georgia.outline, add=TRUE, lwd=2)
title("Geogria with a subset of countues")
```

#### 使い方
1.  地図の描画領域の設定  
2. 領域に対応した地図タイルをダウンロード    
3. そのタイルの上に地図を描画する  

```r
# 地図タイルの左上と右下の点を指定することで領域を設定する
ul <- as.vector(cbind(bbox(georgia.sub)[2,2], bbox(georgia.sub)[1,1]))
lr <- as.vector(cbind(bbox(georgia.sub)[2,1], bbox(georgia.sub)[1,2]))

# 地図タイルをダウンロード
MyMap <- openmap(ul, lr, "mapquest")
```
