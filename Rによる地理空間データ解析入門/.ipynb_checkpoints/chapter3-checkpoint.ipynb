### 3.2.3地図を装飾する

```r
plot(blocks, lwd = 0.5, border="grey50")
plot(breach, add=TRUE, col="red", pch=1)

# 縮尺の追加
map.scale(534750, 152000, miles2ft(2), "Miles", 4, 0.5)

# 北向マークの追加
north.arrow(534750, 154000, miles2ft(0.25), col="lightblue")

# titleの追加
title("New Haven, CT.")
```

### 3.2.4 地図を保存する

```r
# pdfとして保存
png(file="map.png", width = 800, height = 600)

plot(blocks, lwd = 0.5, border="grey50")
plot(breach, add=TRUE, col="red", pch=1)

# 縮尺の追加
map.scale(534750, 152000, miles2ft(2), "Miles", 4, 0.5)

# 北向マークの追加
north.arrow(534750, 154000, miles2ft(0.25), col="lightblue")

# titleの追加
title("New Haven, CT.")

# plotではなくpdfに出力するよう命令
dev.off()
```
