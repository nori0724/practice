## 人気の型

```python
df_type=pd.DataFrame([])
```

```python
df
```

#### ダウン

```python
df_down=df[df["item"].str.contains("ダウン")]
#ファーストダウンのフリースを削除
df_down=df_down.drop(188)
df_down
```

```python
len_down=len(df_down)
df_type["DownJacket"]=[len_down]
```

#### スポーツミックス(ナイロン・トラックジャケット)

```python
df_nairon=df[df["item"].str.contains("ナイロン")]
```

```python
df_track=df[df["item"].str.contains("トラック")]
```

```python
df_sports=pd.concat([df_nairon, df_track]).sort_values("rank").drop_duplicates()
#DIESELナイロンコーチジャケット #nano・universe起毛ナイロンコーチジャケット
df_sports=df_sports.drop(92).drop(173)
df_sports
```

```python
len_sports=len(df_sports)
df_type["SportsMix"]=[len_sports]
```

#### ライダース・レザー

```python
df_riders=df[df["item"].str.contains("ライダース")]
df_riders
```

```python
df_leather=df[df["item"].str.contains("レザー")]
df_leather
```

```python
df_riders=pd.concat([df_riders, df_leather]).sort_values("rank").drop_duplicates()
df_riders
```

```python
len_riders=len(df_riders)
df_type["LeatherRiders"]=[len_riders]
```

#### ボア・フリース

```python
df_boa=df[df["item"].str.contains("ボア")]
df_boa
```

```python
df_fleece=df[df["item"].str.contains("フリース")]
df_fleece
```

```python
df_boa_fleece=pd.concat([df_boa, df_fleece]).sort_values("rank").drop_duplicates()
df_boa_fleece
```

```python
len_boa_fleece=len(df_boa_fleece)
df_type["Fleece"]=[len_boa_fleece]
```

#### マウンテンパーカー

```python
df_mountain=df[df["item"].str.contains("マウンテン")]
df_mountain
```

```python
len_mountain=len(df_mountain)
df_type["Mountain"]=[len_mountain]
```

#### コート

```python
df_coat=df[df["item"].str.contains("コート")]
df_coat
```

```python
len_coat=len(df_coat)
df_type["Coat"]=[len_coat]
```

#### コーチジャケット

```python
df_coachj=df[df["item"].str.contains("コーチジャケット")]
df_coachj
```

```python
len_coachj=len(df_coachj)
df_type["Coach"]=[len_coachj]
```

### 集計

```python
df_type=df_type.T.sort_values(0, ascending = False)
df_type.columns=["count"]
```

```python
df_chester=df_coat[df_coat["item"].str.contains("チェスター")]
df_mod=df_coat[df_coat["item"].str.contains("モッズ")]
df_stainless=df_coat[df_coat["item"].str.contains("ステンカラー")]
df_trench=df_coat[df_coat["item"].str.contains("トレンチ")]
```

```python
df_type_coat=pd.DataFrame([])
df_type_coat["Stainless"]=[len(df_stainless)]

df_type_coat["Chester"]=[ len(df_chester)]

df_type_coat["Trench"]=[ len(df_trench)]

df_type_coat["Mod"]=[ len(df_mod)]

df_type_coat=df_type_coat.T
df_type_coat.columns=["count"]
```
