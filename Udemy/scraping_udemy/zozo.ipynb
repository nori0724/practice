# ZOZOランキングのスクレイピング

```python
rankitemsets=browser.find_elements_by_class_name("rankItemset")

rankitemset_list=[]
for rankitemset in rankitemsets:
    #rankitemsetの中のブランド名を取得
    brandname=rankitemset.find_element_by_tag_name("p").text
    #rankitemsetの中のアイテム名を取得
    itemname=rankitemset.find_element_by_tag_name("dt").text
    #rankitemsetの中の価格を取得
    price=rankitemset.find_element_by_class_name("price")
    taxinprice=price.text.split("税込")[0].strip("¥").replace(",", "")
    discount=price.text.split("税込")[1].strip("(").strip("%OFF)")
    
    rankitemset_list.append([brandname, itemname, taxinprice, discount])
    time.sleep(1)
    
    
```

```python
df=pd.DataFrame(rankitemset_list)
df.columns=["brand", "item", "price", "discount"]
```

```python
df
```

```python
df["discount"] = df["discount"].replace("", 100).astype("int")/100
df["price"] = df["price"].astype("int")
df["originalprice"] = df["price"]/df["discount"]
df["originalprice"] = df["originalprice"].apply(lambda x :'{0:.0f}'.format(x) )
df["discount"] = 1-df["discount"]
```
