```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import mstats
import pystan
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
df=pd.read_excel("./data/multilevel_modeling.xlsx")
```

```python
#１５人の学生あり
df["id"].unique()
```

```python
groups=df.groupby("id")
```

```python
#年齢ごとの身長を描画
#個人ごとの傾きを確認
plt.figure(figsize=(9,9))
for name, group in groups:
    plt.plot(group["age"], group["height"], label=name)
plt.legend()
```

```python
stan_model="""
data{
    int N;
    int N_id;
    real X[N];
    real Y[N];
    int<lower=1, upper=N_id> s_id[N];
}

parameters{
    real a0;
    real b0;
    real a_id[N_id];
    real b_id[N_id];
    real<lower=0> s_a;
    real<lower=0> s_b;
    real<lower=0> s_y;
}

transformed parameters{
    real a[N_id];
    real b[N_id];
    for (n in 1:N_id){
        a[n]=a0+a_id[n];
        b[n]=b0+b_id[n];
    }
}

model{
    for (id in 1:N_id){
        a_id[id] ~ normal(0, s_a);
        b_id[id] ~normal(0, s_b);
    } 
    
    for (n in 1:N){
        Y[n]~normal(a[s_id[n]]*X[n]+b[s_id[n]], s_y);
    }
    

}

"""
```

```python
sm=pystan.StanModel(model_code=stan_model)
```

```python
stan_data={"N":df.shape[0], "N_id":15, "X":df["age"], "Y":df["height"], "s_id":df["id"]}
```

```python
fit=sm.sampling(data=stan_data, iter=1000, chains=3, seed=123)
```

```python
fig=fit.plot()
```

```python
ms_a=fit.extract("a")["a"]
ms_b=fit.extract("b")["b"]
```

```python
x=np.arange(18)
df_b=pd.DataFrame([])
for i in range(18):
    df_b[i]=ms_a[:,0]*x[i]+ms_b[:,0]
```
