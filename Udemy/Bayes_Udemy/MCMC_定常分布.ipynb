```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
plt.style.use("ggplot")
np.random.seed(123)
```

```python
p_trans = np.zeros([3,3])
```

```python
#移動する確率を定義
#0...ofice
#1...kyukei
#2....jikken

p_trans[0,0]=0.1
p_trans[0,1]=0.2
p_trans[0,2]=0.7
p_trans[1,0]=0.1
p_trans[1,1]=0.4
p_trans[1,2]=0.5
p_trans[2,0]=0.3
p_trans[2,1]=0.3
p_trans[2,2]=0.4
```

```python
p_trans
```

```python
NMCS=400
#初期状態(0...office)
c_state=0
c_arr=[c_state]

for i in range(NMCS):
    #0,1,2からランダム1つだけに選ぶ
    current=np.random.choice(3,1, p=p_trans[c_state, :])
    c_state=current[0]
    c_arr.append(c_state)
    
df=pd.DataFrame(c_arr)
```
